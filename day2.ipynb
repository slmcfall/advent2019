```python
def run_program(prgrm):
    # create copy of program to place "in memory"
    prgrm = prgrm.copy()
    
    # iterate over program, by opcode
    for index,opcode in enumerate(prgrm):
        a = index % 4  # check to make sure it's an opcode
        if a == 0:
            if opcode == 1 or opcode == 2:
                # get code positions
                adr_1 = prgrm[index+1]
                adr_2 = prgrm[index+2]
                adr_3 = prgrm[index+3]

                # get position values
                pm_1 = prgrm[adr_1]
                pm_2 = prgrm[adr_2]

                if opcode == 1:
                    # addition
                    prgrm[adr_3] = pm_1 + pm_2

                if opcode == 2:
                    # subtraction
                    prgrm[adr_3] = pm_1 * pm_2
                    
    return (prgrm)
```
